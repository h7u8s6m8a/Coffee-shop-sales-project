# 3 Data Cleaning

### 3.2 Convert Date & Time

```python
df["transaction_date"]=pd.to_datetime(df["transaction_date"], dayfirst=True, errors="coerce")
```

```python
df["transaction_time"]=pd.to_datetime(df["transaction_time"]).dt.time
```

```python
df.head()
```

### Create Total Sales Column

```python
df["Total_sales"] = df["unit_price"]*df["transaction_qty"] 
```

### Feature Engineering

```python
df["day_name_calc"]=df["transaction_date"].dt.day_name()
```

```python
df["month_calc"]=df["transaction_date"].dt.month_name()
```

```python
df["hour_calc"]=pd.to_datetime(df["transaction_time"], format="%H:%M:%S").dt.hour
```

```python
def assign_time_slot(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 22:
        return "Evening"
    else:
        return "Night"
df["time_slot"]= df["hour_calc"].apply(assign_time_slot)
```

```python
df["day_type"]=df["day_name_calc"].apply(lambda x: "Weekend" if x in ["Saturday", "Sunday"] else "Weekday")
```

```python
# Rename duplicates
df = df.rename(columns={
    "month": "month_name",       # if first 'month'
    "month.1": "month_num"     # if second 'month'
})
```

```python
df.columns
```

```python
df=df.drop_duplicates()                      ### DROP DUPLICATES IF EXISTS    
```
